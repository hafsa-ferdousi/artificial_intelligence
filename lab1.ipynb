```python
from queue import Queue
class GridWorldEnv:
    def __init__(self,size,goal,agent_pos):
        self.size=size
        self.goal=goal
        self.agent_pos= agent_pos
        self.visited=[]
    def get_state(self):
        return self.agent_pos

    def is_goal_reached(self):
        return self.agent_pos == self.goal

    def updateAgentPos(self,agentPos):
        self.agent_pos=agentPos

class Agent:
    def __init__(self,env):
        self.env=env

    def perceive(self,current_state,qu):
        x=current_state[0]
        y=current_state[1]
        data=[]
        if(y-1>=0 and [x,y-1] not in self.env.visited and [x,y-1] not in list(qu.queue)):
            data.append([x,y-1])
        if (y + 1 < self.env.size and [x, y + 1] not in self.env.visited and [x,y+1] not in list(qu.queue)):
            data.append([x, y + 1])
        if (x + 1 < self.env.size and [x+1, y ] not in self.env.visited and [x+1,y] not in list(qu.queue)):
            data.append([x+1, y ])
        if (x - 1 >= 0 and [x-1, y ] not in self.env.visited and [x-1,y] not in list(qu.queue)):
            data.append([x-1, y ])
        return data

    def act(self, state):
        self.env.updateAgentPos(state)
        '''if(action=="Up"):
            if(self.env.agent_pos[1]>0):
                self.env.updateAgentPos([self.env.agent_pos[0],self.env.agent_pos[1]-1])
        elif (action == "down"):
            if (self.env.agent_pos[1] < self.env.size):
                self.env.updateAgentPos([self.env.agent_pos[0], self.env.agent_pos[1] + 1])'''

env=GridWorldEnv(4,[3,2],[3,0])
agent=Agent(env)
q = Queue()
q.put(agent.env.get_state())


while not q.empty():
    current_state = q.get()
    print(current_state)
    agent.env.updateAgentPos(current_state)
    agent.env.visited.append(current_state)
    if(agent.env.is_goal_reached()):
        print('Goal')
        break
    percept=agent.perceive(current_state,q)
    for p in percept:
        q.put(p)
```
